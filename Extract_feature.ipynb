```python
#lst là một dynamic array: giá trị chèn vào list lần lượt là
# độ dài dãy con > 3 - giá trị dãy con thuộc (1-9) - vị trí xuất hiện
def tim_cac_quy(arr, n):
  lst = []
  f = np.empty(n, dtype = int)
  f[0] = 1
  
  for i in range(1, n, 1):
    tmp = int(arr[i]) - int(arr[i-1])
    if (tmp == 0):
      f[i] = f[i-1] + 1
      
    else:
      if (f[i-1] > 1):
        lst.append(f[i-1])
        lst.append(int(arr[i-1]))
        lst.append(i-f[i-1])
      f[i] = 1
  if(f[n-1] > 1):
      lst.append(f[n-1])
      lst.append(int(arr[n-1]))
      lst.append(n-f[n-1]) 
  return lst
```

```python
#tìm các sim taxi với cấu trúc ab.ab, ab.ab.ab, abc.abc, abcd.abcd,  ab.ab.ab.ab, abc.abc,abc
#chỉ xét dãy taxi ở cuối 
#thuật toán tham khảo: https://www.baeldung.com/java-repeated-substring
def allCharactersSame(s) :
    n = len(s)
    for i in range(1, n) :
        if s[i] != s[0] :
            return False
 
    return True
 
def tim_sim_taxi(str, ind):
  result = -1
  substr = str[ind:len(str)]
  if allCharactersSame(substr):
    return result 
  new_str = substr + substr;
  cut_new_str = new_str[1:len(new_str)]
  if (cut_new_str.index(substr) != (len(substr)-1)):
    result = 1
  return result
```

```python
#tìm sim taxi dạng ab.ab
df2 = pd.DataFrame()
df2['sim_number'] = df['sim_number']
df2['sim_abab'] = 0
df2['sim_ababab_abcabc'] = 0
df2['sim_abababab_abcdabcd'] = 0
df2['sim_abcabcabc'] = 0
for i in df2.index:
  sim = df2.at[i, 'sim_number']
  if (tim_sim_taxi(sim, len(sim) - 9) == 1):
    df2.at[i, 'sim_abcabcabc'] = 1
  elif (tim_sim_taxi(sim, len(sim) - 8) == 1):
    df2.at[i, 'sim_abababab_abcdabcd'] = 1
  elif (tim_sim_taxi(sim, len(sim) - 6) == 1):
    df2.at[i, 'sim_ababab_abcabc'] = 1
  elif (tim_sim_taxi(sim, len(sim) - 4) == 1):
    df2.at[i, 'sim_abab'] = 1

#trừ đi các sim là sim tứ quý, ngũ quý, lục quý, ... ở đuôi
```

```python
#kiểm tra xem sim có cụm thần tài (39, 79) ở đuôi sim không?
def tim_sim_than_tai(str):
  result = 0
  if ((str.rfind('39') == (len(str)-2)) or (str.rfind('79') == (len(str) - 2))):
    result = 1
  return result

def tim_sim_loc_phat(str):
  result = 0
  if ((str.rfind('68') == (len(str)-2)) or (str.rfind('86') == (len(str) - 2))):
    result = 1
  return result
  
def tim_sim_ong_dia(str):
  result = 0
  if ((str.rfind('38') == (len(str)-2)) or (str.rfind('78') == (len(str) - 2))):
    result = 1
  return result
```

```python
df2['sim_than_tai'] = 0
for i in df2.index:
  if (tim_sim_than_tai(df2.at[i, 'sim_number']) == 1):
    df2.at[i, 'sim_than_tai'] = 1
```

```python
#tìm dãy tiến lên có index là các chữ số cuối, 
#result[0] là dãy tiến lên lớn nhất
# result[1] là vị trí kết thúc dãy tiến lên lớn nhất
# giá sim càng cao nếu dãy tiến lên lớn và các số nằm ở cuối
def tim_day_lien_tiep(arr, n):
  lst = []
  f = np.empty(n, dtype = int)
  f[0] = 1
  
  for i in range(1, n, 1):
    tmp = int(arr[i]) - int(arr[i-1])
    if (tmp == 1):
      f[i] = f[i-1] + 1
      
    else:
      if (f[i-1] > 2):
        lst.append(f[i-1])
        lst.append(int(arr[i-1]))
        lst.append(i-f[i-1])
      f[i] = 1
  if(f[n-1] > 2):
      lst.append(f[n-1])
      lst.append(int(arr[n-1]))
      lst.append(n-f[n-1]) 
  return lst
```

```python
# Đánh chỉ số sim_tien_len là dãy lớn nhất trong sim (dãy ở đuôi)
# Với các sim có dãy tiến lên < 3 loại
# với sim_tien_len_giua chỉ các sim có dãy ở giữa sim, giá trị là dãy lớn nhất tìm được

df1 = pd.DataFrame()
df1['sim_number'] = df['sim_number']
df1['sim_tien_len_1'] = 0
df1['sim_tien_len_2'] = 0
df1['sim_tien_len_3'] = 0
df1['day_tien_len_cuoi'] = 0
length = 9

for i in df.index:
  res = tim_day_lien_tiep(df1.at[i, 'sim_number'], length)
  cnt = 0
  l = len(res)
  for j in range(0, l , 3):
    cnt = cnt + 1
    value = res[j]
    col = 'sim_tien_len_' + str(cnt)
    df1.at[i, col] = value
    if(res[j+2] == length - res[j]):
      df1.at[i, 'day_tien_len_cuoi']  = 1
  
```

```python
#tìm sim gánh đảo
def updateString(xau):
  newStr = '#'
  for i in xau:
    newStr = newStr + i
    newStr = newStr + '#'
  return newStr

def tim_sim_ganh_dao(sim):
  result = np.empty(2, dtype = int)
  n = len(sim)
  maxLength = 1
  start = 0
  for i in range(n):
        for j in range(i, n):
            flag = 1
            # Check palindrome
            rep = 1
            for k in range(0, ((j - i) // 2) + 1):
                if (sim[i + k] != sim[j - k]):
                    flag = 0
                elif ((i+k+1) < n) :
                  if (sim[i+k] != sim[i+k+1]):
                    rep = 0
            # Palindrome
            if rep == 1:
              flag = 0
            if (flag != 0 and (j - i + 1) > maxLength):
                start = i
                maxLength = j - i + 1
  result[0] = maxLength
  result[1] = start           
  return result
```

```python
df4 = pd.DataFrame()
df4['sim_number'] = df['sim_number']
df4['sim_ganh_dao'] = 0
length = 9
for i in df4.index:
  res = tim_sim_ganh_dao(df4.at[i, 'sim_number'])
  if (res[0] >= 4) & ((length - res[0]) == res[1]):
      df4.at[i, 'sim_ganh_dao'] = res[0]
```
